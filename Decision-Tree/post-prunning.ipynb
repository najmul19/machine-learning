```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris # small datsets for post prunning
```

```python
iris = load_iris()
iris
```

```python
iris.data #input features
```

```python
iris.target #output feature
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
#independent and dependent features
X = df.iloc[:,:-1] # all row but last column not include as it is not independet feature
y = iris.target
```

```python
X,y
```

```python
# train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#by deafault it is a post prunning technique because sone of the paramater value none by default
treemodel=DecisionTreeClassifier(max_depth=2)
```

```python
treemodel.fit(X_train,y_train) # model train done
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```

```python
plt.show()
```

```python
#prediction

y_pred=treemodel.predict(X_test)
```

```python
y_pred
```

```python
# check accuracy

from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```
