```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib
```

```python
df = pd.read_csv('homeprices.csv')
```

```python
df.head()
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='blue', marker='*')
```

```python
area = df.drop('price', axis=1)
```

```python
area
```

```python
price=df['price']
price
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
```

```python
model.fit(area,price)
```

```python
model.predict([[3300]])
```

```python
m=model.coef_ #slop
m
```

```python
c = model.intercept_
c
```
