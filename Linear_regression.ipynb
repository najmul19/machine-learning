```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/sample_data/homeprices (1).csv")
```

```python
df.head()
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price, color='red', marker="+")
```

```python
new_df = df.drop("price",axis=1)
#new_df = df.drop("price",axis="columns")
new_df
```

```python
price = df.price
price
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,price)
```

```python
reg.predict([[3300]])
```

```python
m = reg.coef_
m
```

```python
c= reg.intercept_
c
```

```python
3300*m+c
```

```python
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(df.area,df.price, color="red", marker="+")
#plt.plot(df.area, df["price"], color='blue')
plt.plot(df.area, reg.predict(df[["area"]]), color='green')
```

```python
aread_df = pd.read_csv("/content/sample_data/areas (1).csv")
aread_df
```

```python
pred = reg.predict(aread_df)
pred
```

```python
aread_df["price"] = pred
aread_df
```

```python
aread_df.to_csv("house_price_prediction.csv")
```
